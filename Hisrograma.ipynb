```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('./imgs/zelda.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

```

```python
def calculate_hist (image):
  hist = cv2.calcHist([image], [0], None,[256], [0, 256])
  
  plt.plot(hist, color = 'black')
  plt.title('Histograma')
  plt.xlabel('Intensidade')
  plt.ylabel('Quantidade de Pixels')
  plt.show()
```

```python
calculate_hist(image)
```

```python
equalizer_image = cv2.equalizeHist(image)
```

```python
calculate_hist(equalizer_image)
```

```python
plt.figure(figsize= (10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap = 'gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalizer_image, cmap = 'gray')
plt.title('Imagem Equalizada')
plt.axis('off')

plt.show()
```

```python
def threshoting(image, threshold):
  _, thresh = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)
  return thresh
```

```python
threshold_image = threshoting(image, 100)
calculate_hist(threshold_image)
plt.imshow(threshold_image, cmap='gray')
plt.title('Imagem Limiarizada')
```
