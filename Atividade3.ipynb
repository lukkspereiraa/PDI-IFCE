```python
import cv2  #opencv
import matplotlib.pyplot as plt #matplotlib
import numpy as np #numpy
```

```python
imagemAtividade = cv2.imread('./imgs/astroBot.jpg')
imagemAtividadeComvertidaParaRGB = cv2.cvtColor(imagemAtividade, cv2.COLOR_BGR2RGB)
```

```python
def show_image(image, title="Imagem"):
    plt.imshow(image, cmap='gray' if len(image.shape) == 2 else None)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
red, green, blue = cv2.split(imagemAtividadeComvertidaParaRGB)
blank = np.zeros(imagemAtividadeComvertidaParaRGB.shape[:2], dtype='uint8')
canal_vermelho = cv2.merge([red, blank, blank])
canal_verde = cv2.merge([blank, green, blank])
canal_azul = cv2.merge([blank, blank, blue])
```

```python
gray = cv2.cvtColor(imagemAtividadeComvertidaParaRGB, cv2.COLOR_RGB2GRAY)
```

```python
def escolhaQualFiltro():
    while True:
        print("\n" + "-" * 10)
        print("MENU")
        print("-" * 10)
        print("DIGITE 1: VERMELHO")
        print("DIGITE 2: VERDE")
        print("DIGITE 3: AZUL")
        print("DIGITE 4: RGB")
        print("DIGITE 5: PRETO E BRANCO")
        print("DIGITE 0: SAIR DO PROGRAMA")

        try:
            resposta = int(input("Escolha uma opção: "))
            if resposta == 1:
                show_image(canal_vermelho, "Canal Vermelho")
            elif resposta == 2:
                show_image(canal_verde, "Canal Verde")
            elif resposta == 3:
                show_image(canal_azul, "Canal Azul")
            elif resposta == 4:
                show_image(imagemAtividadeComvertidaParaRGB, "Imagem RGB")
            elif resposta == 5:
                show_image(gray, "Imagem Preto e Branco")
            elif resposta == 0:
                print("Saindo do programa.")
                break
            else:
                print("Opção inválida. Tente novamente.")
        except ValueError:
            print("Entrada inválida. Digite um número.")
```

```python
escolhaQualFiltro()
```
