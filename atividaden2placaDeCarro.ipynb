```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = './imgs/car_plate_1.png'  
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(10, 4))
plt.imshow(image, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')
plt.show()
```

```python
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(8, 4))
plt.plot(hist, color='black')
plt.title('Histograma da Imagem')
plt.xlabel('Intensidade de Pixel')
plt.ylabel('Número de Pixels')
plt.show()
```

```python
_, thresh_image = cv2.threshold(image, 118, 250, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(10, 4))
plt.imshow(thresh_image, cmap='gray')
plt.title('Imagem com Threshold')
plt.axis('off')
plt.show()
```

```python
filtered_image = cv2.medianBlur(thresh_image, 3)
```

```python
contours, _ = cv2.findContours(filtered_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
image_contours = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 1)
```

```python
plt.figure(figsize=(10, 4))
plt.imshow(image_contours)
plt.title('Imagem com Contornos Detectados')
plt.axis('off')
plt.show()
```
